# Importing needed libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
```

# Load data

```python
dataset = pd.read_csv("concrete_data.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
dataset.dtypes
```

# Split  data into predictors and target

```python
features = dataset.columns
features
```

```python
X = dataset[features[features != "Strength"]]
y = dataset["Strength"]
```

```python
X.head()
```

```python
y.head()
```

## Data normalization

```python
X = ( X - X.mean() ) / X.std()
X.head()
```

```python
n_cols = X.shape[1]
n_cols
```

# Build a Neural Network

```python
def Regression_model() :
    #create model
    model = Sequential()
    model.add(Dense(50 , activation='relu' , input_shape = (n_cols,)))
    model.add(Dense(30 , activation='relu' , input_shape = (n_cols,)))
    model.add(Dense(1))

    #compile model
    model.compile(optimizer='adam' , loss='mean_squared_error')
    return model
```

# Train and test the Network

```python
x_train , x_test , y_train , y_test = train_test_split(X , y , test_size=0.3 , random_state=4)
```

### Build and fit the model

```python
model = Regression_model()
model.fit(x_train , y_train , validation_split = 0.3 , epochs= 30 , verbose=2)
```

```python
print("Evaluate on test data")
results = model.evaluate(x_test, y_test, batch_size=128)
print("test loss, test acc:", results)
```

### with 100 epoch for training

```python
model.fit(x_train , y_train , validation_split = 0.3 , epochs = 100 , verbose= 2)
```

```python
results = model.evaluate(x_test, y_test, batch_size=128)
print("test loss, test acc:", results)
```
